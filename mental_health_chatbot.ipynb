## 2. Fine-Tune with Hugging Face Trainer


```python
import subprocess, sys
result = subprocess.run([sys.executable, 'train_empathetic.py'], capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print(result.stderr)
else:
    print('Model saved to ./model/')
```

## 3. Test the Chatbot


```python
import sys
sys.path.insert(0, '.')

from chat_app import EmpatheticBot

bot = EmpatheticBot()

queries = [
    'I feel really stressed about exams.',
    'I have been anxious and cannot sleep.',
]

for q in queries:
    print('User:', q)
    print('Bot:', bot.reply(q))
    print('-' * 50)
```
